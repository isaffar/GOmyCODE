```python
def plot_correlation_map( df ): # je n'ai pas compris cette fonction

    corr = df.corr()

    s , ax = plt.subplots( figsize = (12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }
        
    )
plot_correlation_map(df)
```
